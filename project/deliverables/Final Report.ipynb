# Stakeholder Report — Market & Single-Stock Modeling

## Executive Summary
- We modeled MSFT and AAPL daily returns as functions of the market (SPY) using linear regression and tested alternatives (polynomial fit; Gaussian vs bootstrap residual CIs).
- Results are broadly **robust for MSFT** (linear suffices), while **AAPL shows wider residual dispersion** and heavier tails; uncertainty increases when we relax Gaussian assumptions.
- From earlier scenario analysis on preprocessing, reported performance is **sensitive to data-cleaning choices** (mean imputation vs outlier removal).
- **Decision framing:** Prefer the **simple linear model** for MSFT; for AAPL, supplement linear modeling with **bootstrap intervals** and **outlier-aware monitoring**.

## Assumptions & Risks (Clearly Labeled)

**Modeling assumptions**
- Linear relationship between single-stock returns and market return (CAPM-style).
- Residuals are independent and centered at zero; Gaussian errors often assumed for inference.
- Parameters are stable within the training window (stationarity).
- Data preprocessing choices (imputation/outlier handling) do not invert underlying signals.

**Risks / limitations from those assumptions**
- **Linearity risk:** True relation may be mildly nonlinear (tested via polynomial scenario; impact small for MSFT).
- **Distribution risk:** Residuals can be heavy-tailed (AAPL), making Gaussian CIs over-confident.
- **Regime risk:** Betas can drift; results may change across market regimes.
- **Omitted variables:** Single-factor market model excludes sector, macro, or microstructure drivers.

---

## Scenario & Sensitivity Analysis

### A) Model form sensitivity (MSFT)
- **Linear vs Polynomial (degree-2):** Similar scatter around y=x; MAE/RMSE show negligible improvement → **choose linear** for parsimony and stability.

### B) Error-distribution sensitivity (AAPL)
- **Gaussian vs Bootstrap residual mean CI:** Both include zero; **bootstrap is wider**, emphasizing uncertainty under heavy tails.  
- **Implication:** For AAPL, risk estimates and confidence statements should **use bootstrap** (or robust methods) to avoid overstated certainty.

---

## Decision Implications (Actionable Bullets)
- **MSFT:** Use the linear model with standard monitoring; complexity adds little benefit. Deploy with confidence bands from normal theory; keep a watch for parameter drift.  
- **AAPL:** Report uncertainties with **bootstrap CIs**; monitor outliers and subgroup behavior. Avoid over-confidence from Gaussian intervals.  
- **Portfolio-level:** When communicating performance to stakeholders, always pair point estimates with **sensitivity ranges** under alternate assumptions.  
- **Next steps:**  
  1) Add rolling-window beta/MAE diagnostics to detect regime shifts.  
  2) Incorporate robust regression or quantile loss for AAPL to mitigate heavy-tail influence.  
  3) Document preprocessing choices in run configs; rerun sensitivity on release.
